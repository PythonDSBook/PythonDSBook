### コード 3.43 散布図の描画

````python
x1 = np.arange(0, 11)
y1 = np.arange(0, 11)
plt.scatter(x1, y1)
plt.show()
````

### コード 3.44 線グラフの描画

````python
plt.plot(x1, y1)
plt.show()
````

### コード 3.45 曲線のグラフの描画

````python
x2 = np.arange(0, 10.1, 0.1)
y2 = x2 ** 2
plt.plot(x2, y2)
plt.show()
````

### コード 3.46 グラフの情報の描画

````python
plt.plot(x1, y1, label='graph1')
plt.plot(x2, y2, label='graph2')
plt.title('title')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
````
