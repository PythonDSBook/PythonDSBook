## 4.4 データの整形と加工

### コード 4.9 csvファイルの読み込みとデータフレームの作成

```python
URL = 'https://www.nstac.go.jp/sys/files/SSDSE-A-2022.csv'
df = pd.read_csv(URL, skiprows=[0, 1], encoding='shift_jis')
df = df[['都道府県', '市区町村', '総人口', '65歳以上人口']]
```

```python
df
```

### コード 4.10 データフレームの行の並び替え

```python
df.sort_values(['総人口'])
```

```python
df.sort_values(['総人口'], ascending=False)
```

### コード 4.11 データフレームの行の条件抽出

```python
df[(df['総人口'] == 0) | (df['65歳以上人口'] == 0)]
```

### コード 4.12 欠損値を含む行の除去

```python
df.loc[df['総人口'] == 0, '総人口'] = None
df.loc[df['65歳以上人口'] == 0, '65歳以上人口'] = None
df.dropna(inplace=True)
```

```python
df.sort_values(['総人口'])
```

### コード 4.13 データフレームの列の作成

```python
df['高齢化率'] = df['65歳以上人口'] * 100 / df['総人口']
```

### コード 4.14 データフレームの行の並び替え

```python
df.sort_values(['高齢化率'], ascending=False)
```

### コード 4.15 箱ひげ図の描画

```python
plt.boxplot(df['高齢化率'], labels=['高齢化率'])
plt.show()
```

### コード 4.16 外れ値の境界値の計算

```python
c = '高齢化率'
q1 = df[c].quantile(0.25)
q3 = df[c].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
print(f'{lower:.2f}, {upper:.2f}')
```

### コード 4.17 外れ値を含む行の条件抽出

```python
df[(df[c] < lower) | (df[c] > upper)]
```
